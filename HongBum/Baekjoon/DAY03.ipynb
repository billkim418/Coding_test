# 파이썬 기초 - for문

```python
#구구단
number = int(input())
K =1
#range 함수를 이용하여 범위를 지정해줌
for x in range(1,10):
  print("{0} * {1} = {2}".format(number, K, number * K))
  K=K+1
```

```python
# A + B - 3
iter = range(int(input()))
for x in iter:
  first, second = map(int, input().split())
  print(first + second)
```

```python
# 합
n = int(input())
sum = 0
for x in range(n):
  sum = sum + n
  n = n -1
print(sum)
```

```python
#빠른 A + B
iter = range(int(input()))
for x in iter:
  first, second = map(int, input().split())
  print(first + second)
```

```python
#빠른 A + B
##코랩의 경우 sys.stdin으로 인식이 안되는 것을 확인하였음
#구글링 결과 구문은 다 맞고 버전도 다 맞지만 코랩의 경우 인식을 못한다고 함...
import sys
iter = int(input())
for x in range(iter):
  a, b = map(int, sys.stdin.readline().split())
  print(a +b)
```

```python
#N 찍기
number = int(input())
sum = 1
for x in range(number):
  if sum-1 < number:
    print(sum)
  sum = sum +1
```

```python
#기찍 N
number = int(input())
for x in range(number):
  print(number)
  number = number -1
```

```python
#A + B - 7
T = int(input())
start = 1
for x in range(T):
  A, B = map(int, input().split())
  print("Case #{0}: {1}".format(start, A + B))
  start= start + 1
```

```python
#A + B - 8
T = int(input())
start = 1
for x in range(T):
  A, B = map(int, input().split())
  print("Case #{0}: {1} + {2} = {3}".format(start, A, B, A+B))
  start= start + 1
```

```python
# 별 찍기
N = int(input())
count = 1
for x in range(N):
  if count-1 < N:
    print("*" * count)
  count = count +1
```

```python
# 별 찍기 - 오른쪽 정렬
N = int(input())
count = 1
for x in range(N):
  if count-1 < N:
    print(" " * int(N-count)+"*" * count)
  count = count +1
```

```python
# X보다 작은 수
X, Y = map(int, input().split())
Z = list(input().split())
sum = []
for x in range(X):
  if int(Z[x]) <  Y:
    sum.append(Z[x])
print(*sum, sep = " ")
```
