## 꼭 필요한 자료구조 기초
- 탐색(search) : 많은 양의 데이터 중에서 원하는 데이터를 찾는 과정
- 자료구조(Data Structure) : 데이터를 표현하고 관리하고 처리하기 위한 구조
  - 삽입(Push) : 데이터를 삽입한다.
  - 삭제(Pop) : 데이터를 삭제한다.
  - 오버플로(Overflow) : 특정한 자료구조가 수용할 수 있는 데이터의 크기를 이미 가득 찬 상태에서 삽입연산을 수행할 때 발생
  - 언더플로(Underflow) : 특정한 자료구조에 데이터가 전혀 들어 있지 않은 상태에서 삭제 연산을 수행하면 데이터가 전혀 없는 상태가 발생 <br>
- 스택(Stack) : 흔히 선입후출(Frist In Last Out) 구조 또는 후입선출(Last in First Out) 구조라고 부른다.
- 큐(Queue) : 선입선출(First In First Out) 구조이자 '공정한' 자료구조라고 비유된다.

```python
# 5-1 스택(stack) 예제
# keypoint : python에서는 별도의 stack 사용시 선언이 필요없다.

#stack 리스트 선언
stack = []

# sequence : 삽입(5) - 삽입(2) - 삽입(3) - 삽입(7) - 삭제() - 삽입(1) - 삽입(4) - 삭제()
stack.append(5)
stack.append(2)
stack.append(3)
stack.append(7)
stack.pop()
stack.append(1)
stack.append(4)
stack.pop()

#앞의 리스트부터 출력 - 우리가 알고 있는 일반적인 구조
print(stack)
#뒤의 요소부터 출력하고 싶을 경우
print(stack[::-1])
```

```python
#5-2 큐(queue) 예제
# keypoint : 큐 구현을 위해 collections 모듈에서 제공하는 deque 자료구조를 활용한다.
from collections import deque

# 큐(Queue) 구현을 위해 deque 라이브러리 사용
queue = deque()

# sequence : 삽입(5) - 삽입(2) - 삽입(3) - 삽입(7) - 삭제() - 삽입(1) - 삽입(4) - 삭제()
queue.append(5)
queue.append(2)
queue.append(3)
queue.append(7)
queue.popleft() # 가장 먼저 들어온 5 삭제됨
queue.append(1)
queue.append(4)
queue.popleft() # 그 다음으로 들어오 2 삭제됨

# stack과는 다르게 들어온 순서대로 출력됨
print(queue)
# 역순으로 바꾸는 함수
queue.reverse()
# 나중에 들어온 원소부터 출력
print(queue)

#한가지 keypoint -> 자료 변환
# list를 씌워주면 간다히 list형으로 자료구조 변환이 가능하다!
print(list(queue))
```

```python
# 재귀 함수(Recursive Fucntion) : 자기 자신을 다시 호출하는 함수
def recursive_function(i):
  # 종료 조건 명시(100번째 출력)
  if i == 100:
    return
  print(i, '번째 재귀 함수에서', i + 1, '번째 재귀 함수를 호출합니다.')
  recursive_function(i+1)
  print(i, '번째 재귀 함수를 종료합니다.')

recursive_function(1)
```

```python
# 2가지 방식으로 구현한 팩토리얼 예제
# 반복적으로 구현한 n!
def factorial_iterative(n):
  result = 1
  # 1부터 n까지의 수를 차례로 곱하기
  for i in range(1, n+1):
    result *= i
  return result
def factorial_recursive(n):
  if n <= 1:
    return 1
  #n! = n * (n-1)!
  return n * factorial_recursive(n-1)

#두 가지 다른 방식으로 구현한 n! 출력
print('반복적으로 구현', factorial_iterative(5))
print('재귀적으로 구현', factorial_recursive(5))
```
